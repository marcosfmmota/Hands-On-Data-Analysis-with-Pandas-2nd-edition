```python
df = pd.read_csv('data/parsed.csv')
df.headd()
```

## 1. Find the 95 th percentile of earthquake magnitude in Japan using the *mb* magnitude type.

```python
df.loc[(df.parsed_place.str.contains('Japan')) & (df.magType == 'mb'), "mag"].quantile(0.95)
```

## 2. Find the percentage of earthquakes in Indonesia that were coupled with tsunamis.

```python
df['type'].unique()
```

```python
df_indonesia = df[df['parsed_place'] == 'Indonesia']
tsunami_count = df_indonesia[(df_indonesia['tsunami'] == 1) & (df_indonesia['type'] == 'earthquake')].shape[0]
eathquake_count = df_indonesia[df_indonesia['type'] == 'earthquake'].shape[0]
print("Percent: {0:.2f}".format(100 * (tsunami_count/eathquake_count)))
```

## 3. Calculate summary statistics for earthquakes in Nevada.

```python
df[(df.type == 'earthquake') & (df.parsed_place == "Nevada")].describe()
```

## 4. Add a column indicating whether the earthquake happened in a country or US state that is on the Ring of Fire. Use Alaska, Antarctica (look for Antarctic), Bolivia, California, Canada, Chile, Costa Rica, Ecuador, Fiji, Guatemala, Indonesia, Japan, Kermadec Islands, Mexico (be careful not to select New Mexico), New Zealand, Peru, Philippines, Russia, Taiwan, Tonga, and Washington.

```python
df["in_ring_of_fire"] = (df['parsed_place'] == "Alaska")           | \
                        (df['parsed_place'] == "Antarctic")        | \
                        (df['parsed_place'] == "Bolivia")          | \
                        (df['parsed_place'] == "California")       | \
                        (df['parsed_place'] == "Canada")           | \
                        (df['parsed_place'] == "Chile")            | \
                        (df['parsed_place'] == "Costa Rica")       | \
                        (df['parsed_place'] == "Ecuador")          | \
                        (df['parsed_place'] == "Fiji")             | \
                        (df['parsed_place'] == "Guatemala")        | \
                        (df['parsed_place'] == "Indonesia")        | \
                        (df['parsed_place'] == "Japan")            | \
                        (df['parsed_place'] == "Kermadec Islands") | \
                        (df['parsed_place'] == "Mexico")           | \
                        (df['parsed_place'] == "New Zeland")       | \
                        (df['parsed_place'] == "Peru")             | \
                        (df['parsed_place'] == "Philippines")      | \
                        (df['parsed_place'] == "Russia")           | \
                        (df['parsed_place'] == "Taiwan")           | \
                        (df['parsed_place'] == "Tonga")            | \
                        (df['parsed_place'] == "Washington")       
```

## 5. Calculate the number of earthquakes in the Ring of Fire locations and the number outside of them.

```python
df.loc[df["type"] == 'earthquake', "in_ring_of_fire"].value_counts()
```

## 6. Find the tsunami count along the Ring of Fire.

```python
df.loc[(df["type"] == "earthquake") & (df["in_ring_of_fire"]), "tsunami"].sum()
```
