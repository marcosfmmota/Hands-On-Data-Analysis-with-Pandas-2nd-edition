```python
# ITEM d)
df_faang.to_csv('faang.csv', index=False)
```

## Question 6

```python
df_covid = pd.read_csv('covid19_cases.csv')
df_covid.rename(columns={'dateRep':'date'}, inplace=True)
df_covid.loc[:, 'date'] = pd.to_datetime(df_covid.date)
```

```python
df_covid.set_index('date', inplace=True)
df_covid = df_covid.sort_index()
```

```python
df_covid.replace(to_replace={'United_States_of_America': 'USA', 'United_Kingdom':'UK'}, inplace=True)
```

```python
df_filtered = df_covid[df_covid.countriesAndTerritories.isin(["Argentina", "Brazil", "China", "Colombia", "India", "Italy", "Mexico", "Peru", "Russia", "Spain", "Turkey", "UK", "USA"])]
df_filtered.head(10)
```

```python
df_pivoted = df_filtered.pivot(columns='countriesAndTerritories', values='cases').fillna(0)
df_pivoted.head(10)
```
