### 1 - With the earthquakes.csv file, select all the earthquakes in Japan with a magnitude of 4.9 or greater using the mb magnitude type.

```python
earthquakes =  pd.read_csv('earthquakes.csv')
earthquakes.head()
```

```python
earthquakes.query('parsed_place == "Japan" and mag >= 4.9 and magType == "mb"')
```

### 2 - Create bins for each full number of earthquake magnitude (for instance, the first bin is (0, 1], the second is (1, 2], and so on) with the ml magnitude type and count how many are in each bin.

```python
earthquakes_ml = earthquakes.query('magType == "ml"')
pd.cut(
    earthquakes_ml.mag,
    bins=list(range(0, 1 + math.ceil(earthquakes_ml.mag.max()))),
    right=True
).value_counts()
```

### 3 - Using the faang.csv file, group by the ticker and resample to monthly frequency. Make the following aggregations:
    a) Mean of the opening price
    b) Maximum of the high price
    c) Minimum of the low price
    d) Mean of the closing price
    e) Sum of the volume traded

```python
faang = pd.read_csv('faang.csv', index_col='date', parse_dates=True)
faang.head()
```

```python
faang.groupby('ticker').resample('M').agg({
    'open'  :'mean',
    'high'  :'max',
    'low'   :'min',
    'close' :'mean',
    'volume':'sum'
})
```

## 5 - Calculate the rolling 60-day aggregations of the OHLC data by ticker for the FAANG data. Use the same aggregations as exercise 3.

```python
faang.groupby('ticker').rolling('60D').agg({
    'open'  :'mean',
    'high'  :'max',
    'low'   :'min',
    'close' :'mean',
    'volume':'sum'
})
```

## 6 - Create a pivot table of the FAANG data that compares the stocks. Put the ticker in the rows and show the averages of the OHLC and volume traded data.

```python
faang.pivot_table(index='ticker')
```

## 7 -Calculate the Z-scores for each numeric column of Amazon's data (ticker is AMZN) in Q4 2018 using apply() .

```python
faang.query('ticker == "AMZN"')\
     .loc['2018-Q4', ['high', 'low', 'open', 'close', 'volume']]\
     .apply(lambda x: x.sub(x.mean()).div(x.std()))
```

```python
covid = pd.read_csv("covid19_cases.csv")
covid.rename(columns={'dateRep':'date'}, inplace=True)
covid['date'] = pd.to_datetime(covid['date'])
covid = covid.set_index('date')
covid = covid.replace({'United_States_of_America':'USA', 'United_Kingdom':'UK'})
covid = covid.sort_index()
```

```python
covid
```

```python
top_5_countries = covid.groupby('countriesAndTerritories').cases.sum().nlargest(5).index
covid[covid.countriesAndTerritories.isin(top_5_countries)].groupby("countriesAndTerritories").cases.idxmax()
```

```python
top_5_countries
```

```python
covid.groupby(['countriesAndTerritories', pd.Grouper(freq='1D')]).cases.sum()\
    .unstack(0).diff().rolling(7).mean().last('1W')[top_5_countries]
```

```python
covid.reset_index()\
    .pivot(index='date', columns='countriesAndTerritories', values='cases')\
    .drop(columns='China')\
    .fillna(0)\
    .apply(lambda x: x[(x > 0)].idxmin())\
    .sort_values()
```

```python
covid.pivot_table(columns='countriesAndTerritories', values='cases', aggfunc='sum')\
    .T.transform('rank', method='max', pct=True)\
    .sort_values('cases', ascending=False)
```
